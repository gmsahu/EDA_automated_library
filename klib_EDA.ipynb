```python
import klib
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.drop(['alone','adult_male'],axis=1,inplace=True)
```

```python
klib.cat_plot(df) #shows the categorical data plot
```

```python
klib.corr_mat(df) #shows the correlation matrix
```

```python
klib.dist_plot(df) #shows the distribution plot
```

```python
klib.corr_plot(df) #shows the correlation plot
```

```python
klib.missingval_plot(df) #shows the missing values plot
```

```python
df_clean=klib.data_cleaning(df) #cleans the data
```

```python
df_clean.info() #standardise the column name,datatype
```

```python
 klib_clean_columns=klib.clean_column_names(df)  #clean the column names
```

```python
 klib_clean_columns.info()
```

```python
klib_datatype=klib.convert_datatypes(df) # converts existing to more efficient dtypes, also called inside data_cleaning()
```

```python
klib_datatype.info()
```

```python
df_clean.isnull().sum()
```

```python
klib_clean_missing=klib.mv_col_handling(df) #handling missing values
```

```python
klib_clean_missing.isnull().sum()
```
